```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from PIL import ImageTk
from cryptography.hazmat.primitives import serialization
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import hashes

main = Tk()
main.title("Digital Signature")
main.geometry('800x400')


#Create Frame
layoutFrame = Frame(main)
layoutFrame.place(x=640, y=400)

leftFrame = Frame(main,bg='#e5e9fc')
leftFrame.place(x=0, y=0, width=400, height=700)

rightFrame = Frame(main,bg='#B9D3EE')
rightFrame.place(x=370, y=0, width=600, height=700)

#Create labels and entry widgets 
message_label = Label(leftFrame, text='Message:', font=('times new roman', 16, 'bold'))
message_label.grid(row=0, column=0, sticky='w')
message_entry = Entry(leftFrame)
message_entry.grid(row=0, column=1)

user_message_label = Label(rightFrame, text='Received Message:', font=('times new roman', 16, 'bold'))
user_message_label.grid(row=0, column=0, sticky='w')
user_message_entry = Entry(rightFrame)
user_message_entry.grid(row=0, column=1)

signature_received_label = Label(rightFrame, text='Received Signature:', font=('times new roman', 16, 'bold'))
signature_received_label.grid(row=1, column=0, sticky='w')
signature_received_entry = Entry(rightFrame)
signature_received_entry.grid(row=1, column=1)

# Create function to generate user key pair
def generate_user_key_pair():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048,
    )
    return private_key, private_key.public_key()

user_private_key, user_public_key = generate_user_key_pair()

# Create function to sign a message
def sign_message():
    message = message_entry.get()
    if not message:
        messagebox.showerror("Error", "Please enter a message to sign.")
        return
    
    signature = user_private_key.sign(
        message.encode(),
        padding.PSS(
            mgf=padding.MGF1(hashes.SHA256()),
            salt_length=padding.PSS.MAX_LENGTH,
        ),
        hashes.SHA256(),
    )
    
    signature_hex = signature.hex()
    messagebox.showinfo("Signature", f"Signature: {signature_hex}")

# Create function to verify a signature
def verify_signature():
    received_message = user_message_entry.get()
    received_signature_hex = signature_received_entry.get()
    
    if not received_message or not received_signature_hex:
        messagebox.showerror("Error", "Please enter both a received message and a received signature to verify.")
        return
    
    # Convert received signature from hex to bytes
    received_signature = bytes.fromhex(received_signature_hex)
    
    try:
        user_public_key.verify(
            received_signature,
            received_message.encode('utf-8'),
            padding.PSS(
                mgf=padding.MGF1(hashes.SHA256()),
                salt_length=padding.PSS.MAX_LENGTH
            ),
            hashes.SHA256(),
        )
        messagebox.showinfo("Verification", "Signature valid.")
    except Exception as e:
        messagebox.showerror("Verification", f"Signature invalid.\nError: {e}")

# Create buttons 
sign_button = Button(leftFrame, text="Sign Message", command=sign_message)
sign_button.grid(row=3, column=1, columnspan=2, pady=10)

verify_button = Button(rightFrame, text="Verify Signature", command=verify_signature)
verify_button.grid(row=2, column=1, columnspan=2, pady=10)

main.mainloop()
```
